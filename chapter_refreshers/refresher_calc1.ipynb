# Refresher: Calculus 1

## Exercise

Consider the function $f(x)=x^3+1$. HHHHHHAAA!
1. Sketch the graph using a table of values.
2. Determine the $x$-intercept and $y$-intercept of $f$.
3. Where does the graph of $f$ intersect the horizontal line at height $y=9$?
4. Where does the graph o $f$ intersect the graph of the function $x^2+x+1$? 
5. Based on your drawing of the graph $f$, estimate the derivative of $f$ at $1$: $f^\prime(1)$.
6. Based on your drawing of the graph $f$, sketch the derivative $f^\prime$. What could be the functional equation of $f^\prime$?
7. Calculate the functional equation of $f^\prime$.
8. Calculate the slope of the tangent to the graph of $f$ at $x=1$.
9. Determine the functional equation of this tangent. Where does the tangent intersect the $x$-axis, where the $y$-axis?
10. Find the local extrema of $f$. Is it a maximum, minimum, or a saddle point?
11. Estimate the area under the curve of $f$ between $x=0$ and $x=2$ using two bars.
12. Determine the antiderivative of $f$.
13. Calculate the area under the curve of $f$ between $x=0$ and $x=1$ using the fundamental theorem of calculus. 
14. Determine the area enclosed by the graph of $f$, the $x$-axis, and the vertical lines at $x=-2$ and $x=1$.
15. Determine the area enclosed by the graph of $f$ and the graph of $h(x)=4x+1$.
